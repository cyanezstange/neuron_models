```python
import numpy as np
import matplotlib.pyplot as plt
from neuron.spiking import IF

def simple_plot(model):
    plt.figure(figsize=(15,5))
    plt.plot(model.T, model.V, label='Membrane potential')
    plt.plot(model.T, model.I, label='Current')
    plt.title(model.name); plt.xlabel('Time')
    plt.legend(); plt.grid(ls=':'); plt.show()
```

```python
# 1 - Constant current.
model = IF()
model.run(300)

simple_plot(model)
```

```python
# 2 - Variable current.
model = IF()
for _ in range(300):
    model.i = np.sin(model.t)
    model.run_step()

simple_plot(model)
```
